```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
```

Some Data Preprocessing!

1. divides the dataframe into X and Y, which are the independent and dependent variables respectively
2. splits the data in X and Y into a training and testing set where 70% of of the data is allocated to the training set (see train_size = 0.7)
3. normalizes the training and testing independent data vector using the fit_transform function, this function also converts the pandas dataframe into a numpy array
3. convert the dependent variable data into a numpy array

```python
df = pd.read_csv('USA_Housing.csv')
print(df.columns)

X, Y = df.loc[ :, ['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
'Avg. Area Number of Bedrooms', 'Area Population']], df.loc[:, 'Price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

Y_train = Y_train.to_numpy()
Y_test = Y_test.to_numpy()

print(X_train)
print(Y_train)
```

the following cell does the actual gradient descent and regression using sci-kit learn using the SGDRegressor object and returns the coeffiecient and intercept values (w and b in y = w*x + b) in sdgr.coef_ and sgdr.intercept_ fields.

Note that this function performs gradient descent using a method called schochastic gradient descent which is usually much faster than batch gradient descent.

```python
sgdr = SGDRegressor()
sgdr.fit(X_train, Y_train)
print(sgdr)
print(f"number of iterations completed: {sgdr.n_iter_}, number of weight updates: {sgdr.t_}")

print((sgdr.coef_, sgdr.intercept_))
```

prediction of the values using stochastic gradient descent:

```python
Y_predicted = sgdr.predict(X_test)
```

Plot the Y_test and Y_predicted scatterplots against all the 5 features of the dependent variable

```python
n = 5

for i in range(n):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.scatter(X_test[:, i], Y_test, color='r')
    ax.scatter(X_test[:, i], Y_predicted, color='b')
    plt.show()
```
